```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

```python
correlation=df['Open'].corr(df['Close'])
axis_corr=sns.heatmap(corr_matrix,vmin=-1,vmax=1,center=0,cmap=sns.diverging_palette(50, 500, n=500),
square=True)
plt.title('Correlation Matrix of DataFrame')
plt.show()
print(correlation)
```
