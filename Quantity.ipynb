```python
store_sales_pivot = store_sales.pivot(index='DATE', columns='STORE_NBR', values='TotalSales').fillna(0)
display(store_sales_pivot.head())
```

```python
print(f"Number of unique dates in the dataset: {store_sales['DATE'].nunique()}")
```

```python
def calculate_store_similarity(store_sales_df, store1_nbr, store2_nbr):
    """
    Calculates the Pearson correlation between the daily total sales of two stores.

    Args:
        store_sales_df (pd.DataFrame): DataFrame with 'STORE_NBR', 'DATE', 'TotalSales'.
        store1_nbr (int): The store number for the first store.
        store2_nbr (int): The store number for the second store.

    Returns:
        float: Pearson correlation coefficient between the sales of the two stores,
               or None if data is insufficient.
    """
    sales_store1 = store_sales_df[store_sales_df['STORE_NBR'] == store1_nbr].set_index('DATE')['TotalSales']
    sales_store2 = store_sales_df[store_sales_df['STORE_NBR'] == store2_nbr].set_index('DATE')['TotalSales']

    # Align the sales series by date to ensure comparison over the same periods
    aligned_sales = pd.concat([sales_store1, sales_store2], axis=1, keys=[store1_nbr, store2_nbr]).dropna()

    if len(aligned_sales) < 2: # Need at least 2 data points to calculate correlation
        return None

    correlation = aligned_sales[store1_nbr].corr(aligned_sales[store2_nbr])
    return correlation

print("Function 'calculate_store_similarity' re-defined.")
```

```python
trial_store = 77
potential_control_stores = [s for s in store_sales_pivot.columns if s != trial_store]

# Calculate correlations with the trial store
correlations = {}
for store_nbr in potential_control_stores:
    correlation = calculate_store_similarity(store_sales, trial_store, store_nbr)
    if correlation is not None:
        correlations[store_nbr] = correlation

print(f"DEBUG: Number of correlations found: {len(correlations)}") # Added debug print

# Find the control store with the highest correlation
if correlations:
    best_control_store = max(correlations, key=correlations.get)
    max_correlation = correlations[best_control_store]
    print(f"Best control store for trial store {trial_store} is {best_control_store} with correlation {max_correlation:.2f}")
else:
    print(f"Could not find a suitable control store for trial store {trial_store}.")
```
