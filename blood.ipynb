```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Frequency'], bins=20, edgecolor='black') # Using 20 bins as a starting point
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.title('Histogram of Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

class_counts = df['Class'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Class')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define bins for Monetary values
monetary_bins = [0, 500, 1500, 3000, df['Monetary'].max()]
monetary_labels = ['0-500', '501-1500', '1501-3000', '>3000']

# Create a new column with the monetary categories
df['Monetary_Category'] = pd.cut(df['Monetary'], bins=monetary_bins, labels=monetary_labels, right=True, include_lowest=True)

# Get the counts for each category
monetary_category_counts = df['Monetary_Category'].value_counts()

# Create the donut chart
fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(monetary_category_counts,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  pctdistance=0.85)

# Draw a circle in the center to create the donut effect
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

plt.title('Distribution of Monetary Categories (Donut Chart)')
plt.legend(wedges, monetary_category_counts.index, title="Monetary Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
